# Data Cleaning

```python
#Time column is not relevent for our analysis, so dropping that column
df = df.drop('Time',axis=1)
```

```python
#Checking the missing/null values in the dataset
df.isnull().sum()
```

There are some missing values in the dataset

```python
#Removing all the feature with more than 20% missing/null values and imputing the rest with mean of the feature
for col in df.columns:
    if df[col].isnull().sum() > len(df)*0.2:
        df = df.drop(col,axis=1)
    else:
        df[col] = df[col].fillna(df[col].mean())
```

```python
#Dropping the features/columns which have same value for all the rows
for col in df.columns:
    if len(df[col].unique()) == 1:
        df = df.drop(col,axis=1)
```

Since, columns name are not disclosed, and all the missing values are treated. We can go ahead with the further steps as we cannot remove the feature based on visualization.

```python
#Checking the correlation 
plt.figure(figsize=(30,20))
sns.heatmap(df.corr())
```

There are some feature which are strongly correlated. So droping all the feature which have absolute correlation value greater thanm 0.75.

```python
#Removing collinear feature which have absolute correlation value greater than 0.75
def remove_collinear_features(x, threshold):
    
    # Calculate the correlation matrix
    corr_matrix = x.corr()
    iters = range(len(corr_matrix.columns) - 1)
    drop_cols = []

    # Iterate through the correlation matrix and compare correlations
    for i in iters:
        for j in range(i+1):
            item = corr_matrix.iloc[j:(j+1), (i+1):(i+2)]
            col = item.columns
            row = item.index
            val = abs(item.values)

            # If correlation exceeds the threshold
            if val >= threshold:
                # Print the correlated features and the correlation value
                print(col.values[0], "|", row.values[0], "|", round(val[0][0], 2))
                drop_cols.append(col.values[0])

    # Drop one of each pair of correlated columns
    drops = set(drop_cols)
    drops = list(drops)
    print(len(drops))
    x = x.drop(columns=drops,axis=1)
    
    return x
```

```python
df = remove_collinear_features(df,0.75)
```

```python
#converting data type pass/fail to category
df['Pass/Fail'] = df['Pass/Fail'].astype("category")
```

```python
#Boxplot to detect outlier
cols = df.select_dtypes(['int64','float64']).columns
def box_plot(data):
    ncol = 10
    nrow = (len(cols)//ncol)+1
    plt.figure(figsize=(60,60))
    for col, n in zip(cols, np.arange(len(cols))):
        
        ax = plt.subplot(nrow, ncol, n+1)
        sns.boxplot(y=data[col])
```

```python
box_plot(df)
```

We can see that, there are lots of outlier in the dataset.

```python
#Replacing the outlier with median value
for col in df.columns[:-1]:
    q1 = df[col].quantile(0.25)
    q3 = df[col].quantile(0.75)
    iqr = q3 - q1
    low = q1 - (1.5*iqr)
    up = q3 + (1.5*iqr)
    df_in = df[col][(df[col] > low) | (df[col] < up)]
    median = df_in.median()
    df[col] = np.where(df[col] > up, median,df[col])
    df[col] = np.where(df[col] < low, median, df[col])
```

# Data Preprocessing

```python
#Segregating data into independent variable and target variable
x = df.drop('Pass/Fail',axis=1)
y = df['Pass/Fail']
```

```python
#Checking the ditribution training set of target variable
plt.figure(figsize=(20,8))
y.value_counts().plot(kind='pie',autopct = '%0.2f%%')
```

The target variable is highly unbalanced. We need to balanced it

```python
#Balancing the imbalanced data
import imblearn
from imblearn.over_sampling import SMOTE
sm = SMOTE(sampling_strategy='minority')
oversampled_x, oversampled_y = sm.fit_resample(x,y)
```

```python
#Splitting the data into training and testing set
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(oversampled_x,oversampled_y,train_size=0.75,random_state=1)
```

```python
#Sacling the data
from sklearn.preprocessing import StandardScaler
scl = StandardScaler()
x_train_scaled = scl.fit_transform(x_train)
x_test_scaled = scl.transform(x_test)
```

# Comparing all the models

```python
def get_metrics_score(model):
    
    score_list = []
    if model in [log_model_pca,ada_model_pca,rf_model_pca,svc_model_pca,knn_model_pca]:
        pred_train = model.predict(x_train_pca158)
        pred_test = model.predict(x_test_pca158)

        train_acc = round(model.score(x_train_pca158,y_train)*100,2)
        test_acc = round(model.score(x_test_pca158,y_test)*100,2)
    else:
        pred_train = model.predict(x_train_scaled)
        pred_test = model.predict(x_test_scaled)
        train_acc = round(model.score(x_train_scaled,y_train)*100,2)
        test_acc = round(model.score(x_test_scaled,y_test)*100,2)
        
    
    train_recall = round(metrics.recall_score(y_train,pred_train)*100,2)
    test_recall = round(metrics.recall_score(y_test,pred_test)*100,2)
    
    train_precision = round(metrics.precision_score(y_train,pred_train)*100,2)
    test_precision = round(metrics.precision_score(y_test,pred_test)*100,2)
    
    train_f1 = round(metrics.f1_score(y_train,pred_train)*100,2)
    test_f1 = round(metrics.f1_score(y_test,pred_test)*100,2)
    
    kfold = KFold(n_splits=10)
    cvs_k = cross_val_score(model,scl.fit_transform(oversampled_x),oversampled_y,cv=kfold)
    
    skf = StratifiedKFold(n_splits=10)
    cvs_s = cross_val_score(model,scl.fit_transform(oversampled_x),oversampled_y,cv=skf)
    
    Kfold_accuracy = cvs_k.mean()*100
    Kfold_std = cvs_k.std()*100
    Stratified_Kfold_accuracy = cvs_s.mean()*100
    Stratified_Kfold_std = cvs_s.std()*100
    
    score_list.append([train_acc,test_acc,train_recall,test_recall,train_precision,test_precision,train_f1,test_f1,
                       Kfold_accuracy,Kfold_std,Stratified_Kfold_accuracy, Stratified_Kfold_std])
    
    return score_list
```

```python
#Defining the models
from sklearn import metrics
models = [rf_model,rf_estimator_tuned,rf_model_pca,ada_model,Ada_estimator_tuned,ada_model_pca,log_model,
         log_model_tuned,log_model_pca,svc_model,svm_model_tuned,svc_model_pca,knn_model,knn_tuned,knn_model_pca]
#Creating empty list to add train and test results
acc_train = []
acc_test = []
recall_train = []
recall_test = []
precision_train = []
precision_test = []
f1_train = []
f1_test = []
Kfold_Accuracy = []
Kfold_std = []
Skf_accuracy = []
skf_std = []
#looping through the models
for mod in models:
    j = get_metrics_score(mod)
    acc_train.append((j[0][0]))
    acc_test.append((j[0][1]))
    recall_train.append((j[0][2]))
    recall_test.append((j[0][3]))
    precision_train.append((j[0][4]))
    precision_test.append((j[0][5]))
    f1_train.append((j[0][6]))
    f1_test.append((j[0][7]))
    Kfold_Accuracy.append((j[0][8]))
    Kfold_std.append((j[0][9]))
    Skf_accuracy.append((j[0][10]))
    skf_std.append((j[0][11]))
```

```python
#Creating Dataframe
comparison_frame = pd.DataFrame({'Model':['rf_model','rf_estimator_tuned','rf_model_pca','ada_model','Ada_estimator_tuned','ada_model_pca','log_model',
         'log_model_tuned','log_model_pca','svc_model','svm_model_tuned','svc_model_pca','knn_model','knn_tuned','knn_model_pca'],
                                          'Train_Accuracy': acc_train,'Test_Accuracy': acc_test,
                                          'Train_Recall':recall_train,'Test_Recall':recall_test,
                                          'Train_Precision':precision_train,'Test_Precision':precision_test,
                                          'Train_F1':f1_train,
                                          'Test_F1':f1_test,
                                          'Kfold_Accuracy':Kfold_Accuracy,
                                          'Kfold_STD':Kfold_std,
                                          'Stratified Kfold Accuracy': Skf_accuracy,
                                          'Stratified Kfold std':skf_std}) 

#Sorting models in decreasing order of test accuracy
comparison_frame.sort_values(by='Test_Accuracy',ascending=False)
```

We can see that, out of all the models tunned svm model and svm after pca model gives better accuracy also there is no overfitting. Also cross validation score is same for both the models. So we can go ahead with svc_model_pca.
